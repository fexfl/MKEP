```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# b)
def logarithmicWindProfile(z,z_0,u_10,z_r):
    return u_10 * np.log(z / z_0)/np.log(z_r / z_0)

u_10 = 10 #m/s
z_0 = np.array([0.01, 1, 10])*10**(-3) #m
z_r = 10 #m

z = np.linspace(2,50,100)
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.plot(logarithmicWindProfile(z,z_0[0],u_10,z_r), z, label="Logarithmic Wind Profile for z_0 = 10^(-2) mm")
plt.plot(logarithmicWindProfile(z,z_0[1],u_10,z_r), z, label="Logarithmic Wind Profile for z_0 = 1 mm")
plt.plot(logarithmicWindProfile(z,z_0[2],u_10,z_r), z, label="Logarithmic Wind Profile for z_0 = 10 mm")
plt.ylabel('z [m]')
plt.xlabel('u(z) [m/s]')
plt.title('Logarithmic Wind Profile for different surface roughnesses')
plt.legend()
plt.grid(linestyle=":", linewidth=1)
```
